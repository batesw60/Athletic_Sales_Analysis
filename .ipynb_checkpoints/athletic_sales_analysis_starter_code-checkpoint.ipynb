```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
path_2021 = "Resources/athletic_sales_2021.csv"
path_2020 = "Resources/athletic_sales_2020.csv"

df_2021 = pd.read_csv(path_2021)
df_2020 = pd.read_csv(path_2020)
```

```python
# Display the 2020 sales DataFrame
df_2020
```

```python
# Display the 2021 sales DataFrame
df_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.dtypes
```

```python
# Check the 2021 sales data types.
df_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales_df = pd.concat([df_2020, df_2021], ignore_index=True, sort=False)
sales_df = sales_df.reset_index(drop=True)
sales_df.head(10)
```

```python
# Check if any values are null.
sales_df.isnull().values.any()
```

```python
# Check the data type of each column
sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
sales_df.invoice_date = pd.to_datetime(sales_df.invoice_date, format="%m/%d/%y")
sales_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".




# Show the top 5 results.
                            
```

#### Using `pivot_table`

```python
# Create Pivot Table
# Show the number products sold for region, state, and city.
regional_sales_df = sales_df.pivot_table(index=["region", "state", "city"], values="units_sold", aggfunc="sum")
regional_sales_df
```

```python
# Rename the "units_sold" column to "Total_Products_Sold"
regional_sales_df = regional_sales_df.rename(columns={"units_sold": "Total_Products_Sold"})
regional_sales_df.head()
```

```python
# Show the top 5 results.
regional_sales_df = regional_sales_df.sort_values(by=["Total_Products_Sold"], ascending=False)
regional_sales_df.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
sales_df["product"].value_counts()
```

#### Using `groupby`

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
women_pt = filtered_women_df.pivot_table(index=["retailer", "region", "state", "city"], values="units_sold", aggfunc="sum")
women_pt.head()
```

```python
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_pt = women_pt.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
women_pt.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
days_women_sales_df = filtered_women_df.pivot_table(index="invoice_date", values="total_sales", aggfunc="sum")
days_women_sales_df.head()
```

```python
# Optional: Rename the "total_sales" column to "Total Sales"
days_women_sales_df = days_women_sales_df.rename(columns={"total_sales": "Total Sales"})
days_women_sales_df.head()
```

```python
# Show the table.
days_women_sales_df.head()
```

```python
days_women_sales_df.sort_values(by="Total Sales", ascending=False)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
resample_df = days_women_sales_df.resample("D").sum()
resample_df.head()
```

```python
# Sort the resampled pivot table in ascending order on "Total Sales".
resample_df = resample_df.sort_values(by="Total Sales", ascending=False)
resample_df.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
resample_week_df = days_women_sales_df.resample("W").sum()
resample_week_df.head()
```

```python
# Sort the resampled pivot table in ascending order on "Total Sales".
resample_week_df.sort_values(by="Total Sales", ascending=False).head(10)
```
